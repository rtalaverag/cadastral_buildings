## Packages

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
import requests
from zipfile import ZipFile
import os
import feedparser
import unicodedata
```


## Parameters


```python
# Introduce province name
province = "Granada"

# Introduce city name
city = "Granada"

# Cadastral link ATOM
cadastral_link = "http://www.catastro.minhap.es/INSPIRE/buildings/ES.SDGC.bu.atom.xml"
```


## Unicode parameters


```python
def removeAccentMarks(string):
    return ''.join(c for c in unicodedata.normalize('NFD', string)
                  if unicodedata.category(c) != 'Mn')

# Create new to keep original names for Map title
province_clean = removeAccentMarks(province).lower()
city_clean = removeAccentMarks(city).lower()
```


## Parsing cadastral link

See http://www.catastro.minhap.es/webinspire/index.html


```python
feed = feedparser.parse(cadastral_link)
entries = feed.entries
```


## Parsing feed


```python
province_feed = [(lambda e: feedparser.parse(e.link))(e) for e in entries if province_clean in removeAccentMarks(e.title).lower()][0]["entries"]

city_feed = [(lambda e: {"title": e["title"], "zipped_gml": e["link"]})(e) for e in province_feed if f"-{city_clean.upper()}" in e["title"].upper()][0]
```


## Creating a local directory to save the GML file (zip)


```python
data_dir = os.path.join(os.getcwd(),"Downloaded_Data", f"{city_clean.lower()}")

if not os.path.exists(data_dir):
    os.makedirs(data_dir)

zipped_data = os.path.join(data_dir, f"{city_clean.lower()}.zip")
```


## Downloading the GML file (zip)


```python
url = city_feed["zipped_gml"]
r = requests.get(url)

with open (zipped_data, 'wb') as f:
    f.write(r.content)
```


## Creating a GeoDataFrame from building.gml file


```python
# Searching for builing file
gml_file = [f for f in ZipFile(zipped_data).namelist() if "building.gml" in f][0]

# Extract all from the ZipFile
with ZipFile(zipped_data, 'r') as zipObj:
   zipObj.extractall(data_dir)

# Read builiding.gml 
gdf = gpd.read_file(f"{data_dir}/{gml_file}")
```


## Getting the center of the city from the geopandas repository


```python
# Original geocoding = EPSG:4326
center_city_geocoding_response = gpd.tools.geocode(city_clean)["geometry"]

# Reproject
center_city_web_mercator = center_city_geocoding_response.to_crs("EPSG:3857")
```


## City buffer and intersect


```python
# Buffer around city centroid
dist_buffer = 2000
city_buffer = gpd.GeoDataFrame({'geometry': center_city_web_mercator.buffer(dist_buffer, resolution=16)})

# Intersect buildings and buffer
intersect = gpd.overlay(gdf.to_crs(epsg=3857), city_buffer, how='intersection')

# Creating a Year column for quantiles representation
intersect['beginning'] = pd.to_datetime(intersect['beginning'], errors='coerce')
intersect['year'] = intersect['beginning'].dt.year
```


## Plotting city buildings


```python
fig,ax = plt.subplots(figsize=(20,20), facecolor='black') 

plt.suptitle(f"{city}", fontsize=75, color='white', y=0.93)

city_buffer.plot(color='grey', ax=ax, alpha=0.2)

intersect.plot(column="year", ax=ax, alpha=0.9, cmap="YlOrBr_r", scheme='quantiles', k=10,
               legend=True, legend_kwds={'ncol':5, 'labelcolor':'white','fontsize': 20, 'facecolor':'None', 'frameon':False, "fmt": "{:.0f}"})

plt.text(0.15, 0.2, 'Fuente: Catastro', ha='left', va='top', transform=fig.transFigure, color='white', fontsize=15)   # position text relative to Figure
ax.set_axis_off()

leg = ax.get_legend()
leg.set_bbox_to_anchor((1., 0., 0., 0.))

# Optional basemap
#ctx.add_basemap(ax, source=ctx.providers.CartoDB.DarkMatter, alpha=0.9)

fig.savefig(f"{city}.png")  
```
